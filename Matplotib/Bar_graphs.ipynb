```python
import matplotlib.pyplot as plt
import numpy as np

weight = ["40","50","60","70","80"]
height = [140,150,160,170,180]

plt.xlabel("Weight")
plt.ylabel("Height")
plt.bar(weight,height,color = 'blue')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apple","Banana","Mango","Orange"]
myexplode = [0.05,0,0,0]

plt.pie(y,labels = mylabels,explode = myexplode,startangle = 90)
plt.legend(title = "Fruits:")
plt.show() 
```

```python
'''
Write a Python programming to create a pie chart with a title popularity of Languages.
languages: hindi, marathi, tamil, telgu, english,gujrathi
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
'''

import matplotlib.pyplot as plt
import numpy as np

languages = ["hindi", "marathi", "tamil", "telgu", "english","gujrathi"]
Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
myexplode = [0.1,0,0,0,0,0]
mycolors = ['r','y','g','b','c','m']

plt.pie(Popularity,labels = languages,shadow = True,explode = myexplode,colors = mycolors)
plt.legend(title = "Languages:")
plt.show()
```

```python
'''
Plot bar chart of name and precentage
Name: Rudra, Mansvi, Jagruti, Neha, kiran
percentage: 71,60,85,56,90
'''

import matplotlib.pyplot as plt
import numpy as np

Name = ["Rudra", "Mansvi", "Jagruti", "Neha", "kiran"]
percentage = [71,60,85,56,90]

x = Name
y = percentage

plt.title("Bar Chart")
plt.xlabel("Name")
plt.ylabel("Percentage")
plt.bar(x,y,color='blue')
plt.show()
```
