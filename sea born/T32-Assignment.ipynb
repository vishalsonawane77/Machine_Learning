```python
'''
Perform following operations on StudentsPerformance.csv
1) Plot histogram for math score column of given dataframe
2) Draw barplot for gender and math score. Take x=gender,
y=math score; and data=df.
3) Draw barplot for gender and reading score.Take x=gender,
y=reading score, and data=df
4) Draw violin plot violinplot .Take x=&quot;gender&quot;, y=&quot;math score&quot; &amp;
data=df
'''
```

```python
#-> 1)Plot histogram for math score column of given dataframe
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('StudentsPerformance.csv')
plt.title("Histogram")
plt.xlabel("maths score")
plt.ylabel("Freqency")
plt.hist(df["math score"],10)
plt.show()
```

```python
#->2)Draw barplot for gender and math score. Take x=gender,y=math score; and data=df

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
df.loc[df['gender']=="male","gender"] = 0
df.loc[df['gender']=="female","gender"] = 1
sns.barplot(x='gender',y='math score',data=df)
```

```python
#3Draw barplot for gender and reading score.Take x=gender,y=reading score, and data=df

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('StudentsPerformance.csv')
sns.barplot(x='gender',y='reading score',data=df)
```

```python
#4)Draw violin plot violinplot .Take x=gender,y=math score,data=df
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('StudentsPerformance.csv')
sns.violinplot(x='gender',y='reading score',data=df)
```
