```python
import pandas as pd

df = pd.read_csv("titanic_data.csv")
df.drop(["PassengerId","Name","Fare","Cabin"],axis=1).isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())       #median for numerical values
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode)   #mode for categorical values
df.drop(["PassengerId","Name","Fare","Cabin"],axis=1).isnull().sum()
df.drop(["PassengerId","Name","Fare","Cabin"],axis=1)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("titanic_data.csv")
lived = df.loc[df["Survived"] == 1]
dead = df.loc[df["Survived"] == 0]

sns.kdeplot(lived["Age"],color='b',fill=True,label='lived')
sns.kdeplot(dead["Age"],color='g',fill=True,label='dead')

plt.show()

```

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("titanic_data.csv")
lived = df.loc[df["Survived"] == 1]
dead = df.loc[df["Survived"] == 0]

sns.kdeplot(lived["Pclass"],color='g',fill=True,label='lived')
sns.kdeplot(dead["Pclass"],color='r',fill=True,label='dead')
```
